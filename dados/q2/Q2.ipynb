```python
import pandas as pd
```

```python
data = pd.read_fwf('DP.txt', skiprows=4) #Deletando as 4 primeiras linhas.
data
```

```python
data = data.drop(0) #Deletando as linhas 0 e 4.
data = data.drop(4)
data
```

```python
data = data.drop(columns=['&', 'df', 'hf', 'm.1']) #Deletando colunas.
```

```python
data
```

```python
data['subsistema'] = data['ss'] #Substituindo o nome da coluna.
```

```python
data['subsistema'] = data['subsistema'].replace('1', 'seco') #Usando replace para substituição de valores.
data['subsistema'] = data['subsistema'].replace('2', 's')
data['subsistema'] = data['subsistema'].replace('3', 'ne')
data['subsistema'] = data['subsistema'].replace('4', 'n')

```

```python
data['m'] = data['m'].replace('1', ':30') #Reescrevendo os valores para o formato de hora min -> 00:00.
data['m'] = data['m'].replace('0', ':0')
```

```python
data['di'] = data['di'].replace('1', '2021-09-25') #Rescrevendo os dias para o formato padrão de data.
data['di'] = data['di'].replace('2', '2021-09-26') #Ano/mês/dia
data['di'] = data['di'].replace('3', '2021-09-27')
data['di'] = data['di'].replace('4', '2021-09-25')

"""Obs: usei como padrão a dta inicial de 2021-09-25, assim como no enunciado."""
```

```python
data['tempo'] = data['di'] + ' '+  data['hi'] + data['m'] #Somando as colunas para formar uma só que contenha todas as infos.
data['tempo'] 
```

```python
data['dia'] = pd.to_datetime(data['tempo'])
data = data.drop(columns=['ss', 'di', 'hi', 'm', 'tempo'])
data
```
